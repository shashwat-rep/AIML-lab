Using Python language do the exploratory data analysis of dataset imported in the lab 4.

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
# Load the dataset
dataset = pd.read_csv('sample.csv')
```

```python
# Descriptive statistics
print("Descriptive Statistics:\n", dataset.describe())
```

```python
# Checking for missing values
print("\nMissing Values:\n", dataset.isnull().sum())
```

```python
# Correlation matrix
correlation_matrix = dataset.corr(numeric_only=True)
print("\nCorrelation Matrix:\n", correlation_matrix)
```

```python
# Data visualization - pair plot for numerical columns
sns.pairplot(dataset.select_dtypes(include=['float64', 'int64']))
plt.show()
```

```python
# Correlation heatmap
plt.figure(figsize=(10, 8))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', fmt='.2f')
plt.title("Correlation Heatmap")
plt.show()
```

```python
# Distribution of each numerical column
dataset.hist(figsize=(10, 10), bins=30)
plt.show()
```

```python
# Boxplot to check for outliers
plt.figure(figsize=(12, 6))
sns.boxplot(data=dataset.select_dtypes(include=['float64', 'int64']), orient='h')
plt.title("Boxplot of Numerical Columns")
plt.show()
```
